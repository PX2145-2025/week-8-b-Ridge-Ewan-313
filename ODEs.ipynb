````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from scipy.optimize import curve_fit
````

````python
x = np.loadtxt('ex0.txt')
plt.plot(x[:,0],x[:,1],'.')

def exp_decay(t,lam):
  y0 = 100
  return y0*np.exp(-lam*t)

popt, pcov = curve_fit(exp_decay, x[:,0], x[:,1], [1])

plt.plot(x[:,0],exp_decay(x[:,0],*popt))
plt.grid()
print(f'lambda = {popt[0]:.5f} +/- {pcov[0,0]:.5f}')
````

## Exercise 1

 (1). [2] Using the last Python snippet as a guide, solve the ODE for a mass on a spring and plot the result for a suitable range of time.

````python
N = 1000 # Number of points
tmax = 10 # Maximum time
# setting up the time array:
t = np.linspace(0.0, tmax, N)

def f(y, t):
  '''
  return derivatives of the array y defined from the 2nd order ODE for mass on a spring
  '''
  # define constants k and m
  k = 10.0
  m = 1.0
  # the first equation dy[0]/dt and the second equation dy[1]/dt
  return [y[1], -(k/m)*y[0]]

# initial values of y and y'
yinit = [0.1, 0]

y = odeint(f, yinit , t)

# plot the solutions
fig, ax = plt.subplots(2, 1, figsize=(9, 6))
ax[0].plot(t, y[:, 0], 'firebrick')
ax[0].set_ylabel('y')
ax[0].set_xlabel('t')
ax[0].set_title('Solution for y')
ax[0].grid()

ax[1].plot(t, y[:, 1], 'firebrick')
ax[1].set_ylabel('y\'')
ax[1].set_xlabel('t')
ax[1].set_title('Solution for y\'')
ax[1].grid()

fig.tight_layout()
fig.savefig('solutions_for_mass_on_spring.png')
````

## Exercise 2

[7] The driven van der Pol oscillator obeys the ODE:

$$
\frac{d^2y}{dt^2} - m(1 - y^2) \frac{dy}{dt} + y - A \sin(\omega t) = 0
$$

It is an example of a non-linear oscillator and can show mathematically chaotic behaviour.

Write the equation as two coupled first order ODEs and calculate its behaviour.

Plot graphs of $y$ versus $t$ and $y'$ versus $y$. The latter plot is called a "phase space plot" or "phase portrait".
(**Hint**: pick starting parameter values of $m = 8$, $A = 1$ and $\omega = 0.6$ and vary them to get interesting plots. You will need to pick a suitable range of time.)

The two coupled first order ODEs are:
$$
\frac{d^2y}{dt^2} - m(1 - y^2) \frac{dy}{dt} + y = A \sin(\omega t)
$$
$$
\frac{dy}{dt} = y\prime
$$
$$
\frac{d^2y}{dt^2} = A \sin(\omega t) + m(1 - y^2)y\prime - y
$$

````python
N = 100000 # Number of points
tmax = 100 # Maximum time
# setting up the time array:
t = np.linspace(0.0, tmax, N)


# redefine function to allow extra parameters and dampening b
def f(y, t, A, m, w):
  '''
  return derivatives of the array y defined from the 2nd order ODE in question
  takes values for mass, m, angular frequency, w and a constant A
  '''
  # the first equation dy[0]/dt and the second equation dy[1]/dt
  return [y[1], A*np.sin(w*t)+m*(1-y[0]**2)*y[1]-y[0]]

# initial values of y and y', and values for A, m and w
yinit = [0.1, 0.0]
A = 1.0
m = 8.0
w = 0.6


# pass k, m and b as a tuple
y = odeint(f, yinit , t, args=(A, m, w, ))

# plot the solutions
fig, ax = plt.subplots(2, 1, figsize=(9, 6))
ax[0].plot(t, y[:, 0], 'firebrick')
ax[0].set_ylabel('y')
ax[0].set_xlabel('t')
ax[0].set_title('Solution for y')
ax[0].grid()

ax[1].plot(y[len(t)//50:, 0], y[len(t)//50:, 1], 'firebrick')
ax[1].set_ylabel('y\'')
ax[1].set_xlabel('t')
ax[1].set_title('Phase portrait - ignoring first 2% time (so initial conditions don\'t matter)')
ax[1].grid()

fig.suptitle(f'A={A:.3f} m={m:.2f} ω={w:.3f}')
fig.tight_layout()
fig.show()



# repeat 5 times with random constants to see the effects of varying them
for i in range(3):
  A = 4.5 * np.random.rand() + 0.5 # random number between 0.5 and 5
  m = 14 * np.random.rand() + 2 # random number between 2 and 16
  w = 1.5 * np.random.rand() + 0.1 # random number between 0.1 and 1.6

  # pass k, m and b as a tuple
  y = odeint(f, yinit , t, args=(A, m, w, ))

  # plot the solutions
  fig, ax = plt.subplots(2, 1, figsize=(9, 6))
  ax[0].plot(t, y[:, 0], 'firebrick')
  ax[0].set_ylabel('y')
  ax[0].set_xlabel('t')
  ax[0].set_title('Solution for y')
  ax[0].grid()

  ax[1].plot(y[len(t)//10:, 0], y[len(t)//10:, 1], 'firebrick')
  ax[1].set_ylabel('y\'')
  ax[1].set_xlabel('t')
  ax[1].set_title('Phase portrait - ignoring first 10% time')
  ax[1].grid()

  fig.suptitle(f'A={A:.3f} m={m:.2f} ω={w:.3f}')
  fig.tight_layout()
  fig.show()
````
